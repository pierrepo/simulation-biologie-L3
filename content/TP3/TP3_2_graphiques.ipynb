## Nuage de points

Pour un exemple fictif, on a mesuré la concentration d'un produit dans le sang (exprimé en mg/L) à différents instants (exprimés en heures).

Voici les valeurs mesurées :

| Temps (h) | Concentration (mg/L) |
|:---------:|:--------------------:|
|     1     |          3.5         |
|     2     |          5.8         |
|     3     |          9.1         |
|     4     |         11.8         |
|     6     |         17.5         |
|     7     |         21.3         |
|     9     |         26.8         |

On souhaite représenter graphiquement l'évolution de la concentration du produit en fonction du temps.

Créons pour cela deux listes, `temps` et `concentrations` qui vont contenir ces valeurs :

````python
temps = [1, 2, 3, 4, 6, 7, 9]
concentrations = [5.5, 7.2, 11.8, 13.6, 19.1, 21.7, 29.4]
````

Chaque couple de valeurs pris dans les listes `temps` et `concentrations` va constituer les coordonnées des points que nous allons afficher dans le graphique. Ainsi, le premier point aura comme coordonnées `{1, 5.5}`, le deuxième `{2, 7.2}`, le troisième `{3, 11.8}`...

Créons maintenant le graphique avec la bibliothèque `matplotlib` :

````python
fig, ax = plt.subplots()
ax.scatter(temps, concentrations, marker="o", color="blue")
ax.set_ylim(0, 40)
ax.set_xlabel("Temps (h)")
ax.set_ylabel("Concentration (mg/L)")
ax.set_title("Concentration de produit en fonction du temps")
plt.show()
````

Explications :

- `plt.subplots()` initialise le graphique.
- `ax.scatter()` dessine les points (*scatter*). On fournit à cette fonction d'abord les abscisses (`temps`) puis les ordonnées (`concentrations`) des points que l'on souhaite représenter. Les arguments `marker` et `color` configurent le type de points à afficher et la couleur de ces points. Matplotlib accepte de nombreux [types de points](https://matplotlib.org/stable/api/markers_api.html) et [couleurs](https://matplotlib.org/stable/gallery/color/named_colors.html#css-colors).
- `ax.set_ylim()` précise l'intervalle de valeurs minimales et maximales pour l'axe des ordonnées.
- `ax.set_xlabel()`, `ax.set_ylabel()` et `ax.set_title()` définissent les étiquettes des axes des abscisses et des ordonnées, ainsi que le titre du graphique.
- Enfin, `plt.show()` affiche le graphique dans le *notebook*.

## Lignes

On modélise l'évolution de la concentration en produit dans le sang en fonction du temps par l'équation :

$$
{\rm concentration} = 3 \times {\rm temps} + 2
$$

On calcule ensuite la concentration théorique en fonction du temps avec cette formule et une boucle `for` :

````python
concentrations_theoriques = []
for t in temps:
    c_theo = 3 * t + 2
    concentrations_theoriques.append(c_theo)
````

Affichons maintenant, dans le même graphique, l'évolution de la concentration expérimentale (en nuage de points) et de la concentration théorique (en ligne) en fonction du temps :

````python
fig, ax = plt.subplots()
ax.scatter(temps, concentrations, marker="o", color="blue")
ax.plot(temps, concentrations_theoriques, ls="-", color="green")
ax.set_ylim(0, 40)
ax.set_xlabel("Temps (h)")
ax.set_ylabel("Concentration (mg/L)")
ax.set_title("Concentration de produit en fonction du temps")
plt.show()
````

Explications :

- `ax.plot()` permet de dessiner des points reliés par des lignes (sans afficher les différents points). Comme avec `ax.scatter()`, on fournit à cette fonction d’abord les abscisses (`temps`) puis les ordonnées (`concentrations_theoriques`) des points que l’on souhaite représenter. Les arguments `ls` et `color` déterminent le [type de ligne](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html) à afficher et la couleur de cette ligne.

Pour améliorer ce graphique, on ajoute une légende :

````python
fig, ax = plt.subplots()
ax.scatter(temps, concentrations, marker="o", color="blue", label="mesures")
ax.plot(temps, concentrations_theoriques, ls="-", color="green", label="modèle")
ax.set_ylim(0, 40)
ax.set_xlabel("Temps (h)")
ax.set_ylabel("Concentration (mg/L)")
ax.set_title("Concentration de produit en fonction du temps")
ax.legend()
plt.show()
````
