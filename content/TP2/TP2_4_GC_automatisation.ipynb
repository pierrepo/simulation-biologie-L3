## Comparaison de modèles pour calculer la température de fusion d'une amorce

La formule que vous avez utilisée pour calculer la température de fusion est :

$$
Tm = 64.9 + 0.41 \times \%GC - \frac{672.4}{nbA + nbT + nbC + nbG}
$$

Il existe une formule plus simple, appelée formule de Wallace, qui n'utilise pas le %GC :

$$
Tm = 2 (nbA + nbT) + 4 (nbG + nbC)
$$

Dans la cellule de code ci-dessous, calculez la température de fusion `Tmw` avec la formule de Wallace pour la séquence précédente (réutilisez les variables `nb_A`, `nb_T`, `nb_C`, `nb_G`) :

Voici un exemple de code possible :

```python
sequence = "AGTCCTGACCTGTCCCG"

nb_A = 0
nb_T = 0
nb_C = 0
nb_G = 0

for base in sequence:
    if base == "A":
        nb_A = nb_A + 1
    elif base == "T":
        nb_T = nb_T + 1
    elif base == "C":
        nb_C = nb_C + 1
    elif base == "G":
        nb_G = nb_G + 1

GC = (nb_G + nb_C)/(nb_A + nb_T + nb_C + nb_G) * 100

Tm = 64.9 + 0.41 * GC - 672.4/(nb_A + nb_T + nb_C + nb_G)
Tmw = 2 * (nb_A + nb_T) + 4 * (nb_G + nb_C)

print(f"Taille de la séquence : {len(sequence)}")
print(f"Température de fusion (formule 1) : {Tm:.2f} °C")
print(f"Température de fusion (formule de Wallace) : {Tmw} °C")
```

Résultats attendus :

```none
Taille de la séquence : 17
Température de fusion (formule 1) : 51.88 °C
Température de fusion (formule de Wallace) : 56 °C
```

Commentaires :

- Les deux températures de fusion obtenues sont proches et compatibles avec l'intervalle de températures de la phase d'hybridation.
- La formule de Wallace ne s'applique en réalité que sur de courtes séquences d'amorce (entre 14 et 20 bases). Au-delà, elle n'est plus fiable, d'où le résultat aberrant obtenu précédemment avec une amorce de séquence beaucoup plus longue (35 bases).
- Les deux formules proposées dans le cadre de ce TP ne font qu'approximer la température de fusion réelle d'une amorce. Il existe des formules plus complexes, qui prennent par exemple en compte la concentration d'amorce et la concentration de sel dans la solution.
