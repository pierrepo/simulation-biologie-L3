# Discrétisation de fonctions mathématiques

## Qu'est-ce que discrétisatiser une fonction ?

En mathématiques, une fonction $f(x)$ est souvent définie sur un **intervalle continu**.  
En informatique, on ne peut travailler, le plus souvent, qu'avec un nombre **fini** de valeurs.

**Discrétiser une fonction** consiste à représenter une fonction continue par un ensemble **fini ($n$) de points** :

$$
\{x_1, f(x_1)\}, \{x_2, f(x_2)\} \dots \{x_n, f(x_n)\}
$$

où les ordonnées $x_i$ sont réparties uniformément sur l'intervalle $[a, b]$ avec un pas entre deux points successifs :

$$
p = \frac{b - a}{n}
$$

## Fonction sinus

Pour vous en rendre compte, on va représenter graphiquement la fonction $\sin$ sur l'intervalle $[0, 4\pi]$ (c'est-à-dire 2 tours).

Chargeons d'abord la bibliothèque `matplotlib` :

````python
import matplotlib.pyplot as plt
%matplotlib inline
````

Créons une fonction qui :

- Définit les abscisses et les ordonnées sous forme de 2 listes.
- Uilise le module `math` de Python qui donne accès à la fonction $\sin$ et à la constante $\pi$.
- Crée le graphique de $\sin(x)$ en fonction de $x$

````python
import math

def discretise_sin(a, b, n):
    # On définit le pas, c'est-à-dire la progression entre les valeurs successives de x.
    p = (b - a) / n
    # On crée deux listes vides pour les abscisses et les ordonnées.
    abscisses = []
    ordonnees = []
    # On calcule x et y pour n points.
    for i in range(n):
        x = a + p * i
        y = math.sin(x)
        abscisses.append(x)
        ordonnees.append(y)
    # On crée le graphique.
    fig, ax = plt.subplots()
    ax.plot(abscisses, ordonnees, color="blue")
    ax.set_xlabel("x")
    ax.set_ylabel("sin(x)")
    ax.set_title(f"sin(x) en fonction de x pour {n} points")
    plt.show()
````

Affichons la discrétisation de la fonction $\sin$ entre $0$ et $4 \pi$ pour 10 points :

````python
discretise_sin(0, 4 * math.pi, 10)
````

Et la même chose pour 100 points :

````python
discretise_sin(0, 4 * math.pi, 100)
````

Plus on ajoute de points, plus ceux-ci vont être rapprochés et plus la représentation graphique sera proche de ce qu'on peut attendre d'une fonction sinus.

## Fonction $e^{-x^2}$

De la même manière, on peut discrétiser la fonction mathématique $e^{-x^2}$, par exemple sur l'intervalle $[-3 ; 3]$

En Python, la fonction exponentielle ($e$) en Python est fournie par la fonction `math.exp()` du module `math`.

Créons cette fois une fonction qui discrétise la fonction $e^{-x^2}$ mais ne s'occupe pas de l'affichage :

````python
import math

def discretise_exp(a, b, n):
    # On définit le pas, c'est-à-dire la progression entre les valeurs de x.
    p = (b - a) / n
    # On crée deux listes vides pour les abscisses et les ordonnées.
    abscisses = []
    ordonnees = []
    # On calcule x et y pour n points.
    for i in range(n):
        x = a + p * i
        y = math.exp(-x*x)
        abscisses.append(x)
        ordonnees.append(y)
    return abscisses, ordonnees
````

Appelons cette fonction pour discrétiser la fonction $e$ avec différents nombres de point, mais toujours dans l'intervalle $[-3 ; 3]$. Affichons ensuite ces points sous forme de lignes.

````python
# On crée les différentes discrétisations
x_6, y_6 = discretise_exp(-3, 3, 6)
x_20, y_20 = discretise_exp(-3, 3, 20)
x_100, y_100 = discretise_exp(-3, 3, 100)

fig, ax = plt.subplots()
ax.plot(x_6, y_6, color="blue", label="6 points")
ax.plot(x_20, y_20, color="royalblue", label="20 points")
ax.plot(x_100, y_100, color="dodgerblue", label="100 points")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.legend()
plt.show()
````

:::{admonition} Remarques
:class: tip

Plus on discrétise une fonction avec un nombre de points élevé, plus la fonction apparait comme « lisse », c'est-à-dire proche de son graphe habituel, mais plus le temps de calcul pour déterminer les coordonnées des points sera important.

:::
