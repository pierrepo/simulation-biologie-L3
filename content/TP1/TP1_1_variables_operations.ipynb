## 1. Définition

La **division entière** est une opération mathématique qui consiste à diviser un nombre entier (le dividende)  par un autre (le diviseur) **sans utiliser de nombres décimaux**.

Elle permet d’obtenir deux résultats :

* **le quotient** : combien de fois le diviseur rentre dans le dividende ;
* **le reste** : ce qui reste après avoir fait ces divisions complètes.

On peut résumer cela par la formule suivante :

$$
dividende = diviseur \times quotient + reste
$$

avec la condition :

$$
0 \le reste < diviseur
$$

Dans une division entière, on distingue donc quatre éléments :

1. **Dividende** : nombre que l'on divise
2. **Diviseur** : nombre par lequel on divise
3. **Quotient** : nombre de divisions complètes possibles
4. **Reste** : ce qui reste après ces divisions

## 2. Exemple

Soit la division :

$$
17 \div 5,
$$

on a alors :

* dividende : 17
* diviseur : 5
* quotient : 3
* reste : 2

Pourquoi ?

$$
5 \times 3 = 15
$$

et

$$
17 - 15 = 2
$$

Donc :

$$
17 = 5 \times 3 + 2
$$

## 3. Avec Python

Python possède deux opérateurs utiles pour la division entière :

| Opération        | Symbole | Signification       |
| ---------------- | ------- | ------------------- |
| division entière | `//`    | renvoie le quotient |
| modulo           | `%`     | renvoie le reste    |

En reprenant l'exemple précédent :

```python
17 // 5
```

résultat :

```none
3
```

Et :

```python
17 % 5
```

résultat :

```none
2
```

Donc :

```python
17 = 5 * (17 // 5) + (17 % 5)
```

On utilise les propriétés de la division entière pour convertir un temps d'incubation en heures et minutes ou déterminer le nombre de codons dans une séquence d'ADN.

## Conversion de types

La conversion de type permet de passer d'un type de variable à l'autre.

On utilise pour cela les fonctions :

- `int()` qui essaie de convertir en entier ;
- `float()` qui essaie de convertir en réel ;
- et `str()` qui essaie de convertir en chaînes de caractères.

````python
int(3.14)
````

````python
int("-45")
````

Les conversions entre types doivent avoir un sens.

Par exemple, on ne peut pas convertir la chaîne de caractères `ATCG` en entier :

````python
int("ATCG")
````

Mais, on peut convertir un nombre entier en réel :

````python
float(5)
````

On peut aussi convertir la chaîne de caractères composée des caractères `3`, `.`, `1` et `4` en réel :

````python
float("3.14")
````

📝 Rappel : attention au séparateur décimal en Python !

Le code suivant va déclencher une erreur :

````python
float("3,14")
````

Pour terminer, on peut convertir n'importe quel nombre (entier, réel) en chaîne de caractères :

````python
str(42)
````

````python
str(3.14)
````

#### Celsius vers Kelvin

Pour passer d'une température en Celsius (°C) à une température en Kelvin (K), on utilise la formule suivante :

$$T_K = T_C + 273,15$$

où $T_K$ représente la température en Kelvin et $T_C$ la température en Celsius.

Dans la cellule suivante :

- Créez une variable `temp_C` qui correspond à une température en Celsius de 20,5 °C.
- Calculez la température correspondante en Kelvin en utilisant la formule précédente et stockez cette valeur dans la variable `temp_K`.
- Affichez la valeur de la variable `temp_K`.

:::{admonition} Éléments de réponse
:class: tip
:class: dropdown

Une réponse possible est :

```python

temp_K = temp_C + 273.15
temp_K
```

Soit 293,65 K.

:class: note

#### Fahrenheit vers Celsius

Pour passer d'une température en Fahrenheit (°F) à une température en Celsius (°C), on utilise la formule suivante :

$$T_C = (T_F - 32) \times \frac{5}{9}$$

où $T_C$ représente la température en Celsius et $T_F$ la température en Fahrenheit.

Dans la cellule suivante :

- Créez une variable `temp_F` qui correspond à une température en Fahrenheit de 98,6 °F.
- Calculez la température correspondante en Celsius en utilisant la formule précédente et stockez cette valeur dans la variable `temp_C`.
- Affichez la valeur de la variable `temp_C`.

:::{admonition} Éléments de réponse
:class: tip
:class: dropdown

Une solution possible est :

```python
temp_F = 98.6
temp_C = (temp_F - 32) * 5/9
temp_C
```

Soit 37,0 °C, qui correspond, approximativement, à la température du corps humain.

::::{admonition} Remarques
:class: note

- Faites attention au séparateur décimal lors de la création de la variable `temp_F`.
- Respectez la priorité des opérations en mettant entre parenthèses l'opération `temp_F - 32`.
