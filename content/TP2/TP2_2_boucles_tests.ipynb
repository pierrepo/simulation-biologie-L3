### Itération sur les éléments d'une liste

Voici les ingrédients que j'ai dans mon placard, exprimés sous forme d'une liste :

````python
placard = ["farine", "oeufs",  "sucre", "lait"]
````

Pour afficher, les uns après les autres, les ingrédients de mon placard, nous pourrions écrire :

````python
print(placard[0])
print(placard[1])
print(placard[2])
print(placard[3])
````

Cette méthode est répétitive et serait très pénible si la liste `placard` contenait 10, 100, 1000 éléments (oui, j'ai de grands placards).

Les boucles ont justement été créées pour automatiser une action (ici afficher chacun des éléments d'une liste) :

````python
for ingredient in placard:
    print(ingredient)
````

Remarques :

- La variable `ingredient` est appelée *variable d'itération* et change de valeur à chaque itération (à chaque répétition) de la boucle. La variable `ingredient` va prendre comme valeur chaque élément de la liste `placard`.
- La ligne débutant par `for` se termine toujours par `:`
- Le bloc d'instructions `print(ingredient)` est **indenté**, c'est-à-dire qu'il est décalé vers la droite de quatre espaces.

Un bloc d'instructions peut contenir plusieurs lignes d'instructions :

````python
placard = ["farine", "oeufs",  "sucre", "lait"]

for ingredient in placard:
    print("J'ajoute un ingrédient :")
    print(ingredient)
print("La pâte à crêpes est prête !")
````

Ici, le bloc d'instructions de la boucle `for` est composé de deux instructions :
```
print("J'ajoute un ingrédient :")
print(ingredient)
```

L'instruction `print("La pâte à crêpes est prête !")` est en dehors du bloc d'instructions de la boucle. Elle est donc exécutée après la boucle. Visuellement, on s'en rend compte, car elle est alignée avec le début de la boucle `for`.

### Itération sur les caractères d'une chaîne de caractères

Souvenez-vous, une chaîne de caractères fonctionne *presque* comme une liste.

On peut parcourir chaque caractère les uns après les autres et les afficher.

````python
sequence = "ATCG"

for base in sequence:
    print(base)
````

On peut aussi utiliser une *f-string* pour rendre l'affichage plus joli :

````python
sequence = "ATCG"

for base in sequence:
    print(f"La base est : {base}")
````

## Tests

Ces notions sont abordées en détail ici : https://python.sdv.u-paris.fr/06_tests/

Les tests sont utilisés pour prendre des décisions.

````python
nombre = 2
````

````python
if nombre == 2:
    print("Gagné !")
````

Remarques :

- La ligne débutant par  `if` se termine toujours par `:`
- `==` (deux fois le symbole `=`) est un opérateur de comparaison d'égalité. On vérifie si `nombre` est égale à `2`. Ne confondez par cet opérateur d'égalité avec l'opérateur d'affectation `=` qui permet de donner une valeur à une variable (par exemple : `a = 1.8`).
- Un bloc d'instructions suit la ligne contenant `if` et est indenté, c'est-à-dire qu'il est décalé vers la droite de quatre espaces.

### Tests à deux cas

````python
nombre = 2

if nombre == 2:
    print("Gagné !")
else:
    print("Perdu !")
````

Remarques :

- Le symbole `:` se trouve à la fin de la ligne débutant par `if` et de celle débutant par `else`
- Il y a un bloc d'instructions (indenté) après la ligne avec le `if`
- Il y a un bloc d'instructions (indenté) après la ligne avec le `else`

### Tests à plusieurs cas

````python
base = "T"
````

````python
if base == "A":
    print("Choix d'une adénine")
elif base == "T":
    print("Choix d'une thymine")
elif base == "C":
    print("Choix d'une cytosine")
elif base == "G":
    print("Choix d'une guanine")
````

On peut également définir un cas « par défaut » avec `else` :

````python
base = "P"

if base == "A":
    print("Choix d'une adénine")
elif base == "T":
    print("Choix d'une thymine")
elif base == "C":
    print("Choix d'une cytosine")
elif base == "G":
    print("Choix d'une guanine")
else:
    print("Révise ta biologie !")
````

## Boucle + tests = automatisation = 🚀

L'utilisation combinée de boucles et de tests permet d'automatiser un traitement de données et d'adapter le comportement du programme en fonction de chaque donnée à traiter.

Cela permet de créer des algorithmes assez sophistiqués.

````python
sequence = "ATTGCGCA"

for base in sequence:
    if base == "A":
        print("La base est une adénine")
    elif base == "T":
        print("La base est une thymine")
    elif base == "C":
        print("La base est une cytosine")
    elif base == "G":
        print("La base est une guanine")
````

Remarques :

- Le premier niveau d'indentation concerne la boucle `for`.
- Le second niveau d'indentation concerne les tests (`if` et `elif`).

### Attention à l'indentation !

Observez avec attention les exemples de code ci-dessous, exécutez-les puis essayez de comprendre ce qui se passe pour la dernière ligne.

````python
nombres = [4, 5, 6]

for nb in nombres:
    if nb == 5:
        print("Le test est vrai")
        print(f"car la variable nb vaut {nb}")
````

````python
nombres = [4, 5, 6]

for nb in nombres:
    if nb == 5:
        print("Le test est vrai")
    print(f"car la variable nb vaut {nb}")
````

:::{admonition} Explications
:class: note
:class: dropdown

La seule différence entre les deux cellules de code ci-dessus est l'indentation de la dernière ligne : `print(f"car la variable nb vaut {nb}")`

Dans la première cellule de code, la dernière ligne est indentée de la même manière que la ligne `print("Le test est vrai")`. Elle ne sera exécutée que si le test `if nb == 5:` est vrai. On obtiendra donc comme résultat :

```none
Le test est vrai
car la variable nb vaut 5
```

Dans la seconde cellule de code, la dernière ligne est indentée de la manière que la ligne `if nb == 5:`. Elle sera exécutée **à chaque iteration** de la boucle `for`, comme le test `if nb == 5:`. On obtiendra alors comme résultat :

```none
car la variable nb vaut 4
Le test est vrai
car la variable nb vaut 5
car la variable nb vaut 6
```

### Compter les bases

Si on peut compter un acide aminé particulier dans une séquence de protéine, on peut aussi compter les différentes bases d'une séquence d'ADN :

````python
ADN = "ATAAAGCCGTTATGCAYCCA"

nb_A = 0
nb_T = 0
nb_C = 0
nb_G = 0

for base in ADN:
    if base == "A":
        nb_A = nb_A + 1
    elif base == "T":
        nb_T = nb_T + 1
    elif base == "C":
        nb_C = nb_C + 1
    elif base == "G":
        nb_G = nb_G + 1

print(nb_A, nb_T, nb_C, nb_G)
````

:::{note}

L'algorithme pour compter un acide aminé ou des bases dans une séquence est identique. Dans les deux cas, on parcourt les acides aminés ou les bases de la séquence (de protéine ou d'ADN) et on compte un acide aminé ou une base de plus à chaque fois que l'acide aminé ou la base recherché·e est identifié·e.
