## Lire un fichier

Depuis l'interface JupyterLab, n'hésitez pas à ouvrir le fichier `zoo.txt` en double-cliquant sur son nom dans l'explorateur de fichiers (*File Browser*, à gauche).

Nous allons maitenant lire le contenu de ce fichier avec Python :

````python
with open("zoo.txt", "r") as input_file:
    for line in input_file:
        print(line)
````

Remarques :

- L'instruction `with` introduit un bloc d'instructions. C'est à l'intérieur de ce bloc que nous effectuons toutes les opérations sur le fichier. L'instruction `open()` ouvre le fichier. Une fois sorti de ce bloc d'instructions, le fichier est automatiquement fermé.
- La variable `input_file` est une variable qui permet d'accéder au contenu du fichier depuis Python.
- Le mode `"r"` (pour *read*) indique qu'on lit le fichier.
- L'instruction `for` permet d'itérer (c'est-à-dire de parcourir) chaque ligne du fichier, une par une.
- Il y a une ligne vide entre chaque animal car la ligne contient elle-même un saut de ligne qui s'additionne à celui de l'instruction `print()`. Vous visualiserez ce saut de ligne un peu plus bas dans le notebook.

Pour supprimer le saut de ligne supplémentaire, on peut « nettoyer » la ligne lue dans le fichier avec  la méthode `.strip()` :

````python
with open("zoo.txt", "r") as input_file:
    for line in input_file:
        print(line.strip())
````

C'est beaucoup mieux ! Nous affichons maintenant le contenu du fichier `zoo.txt` tel qu'il apparait dans un éditeur de texte, avec un nom d'animal par ligne.

## Lire un fichier en une seule fois

Il est parfois utile de lire le contenu d'un fichier en une seule fois. C'est ce que propose la méthode `.read()` :

````python
with open("zoo.txt", "r") as input_file:
    content = input_file.read()
````

L'entièreté du fichier `zoo.txt` est stocké dans la variable `content` que vous pouvez afficher en exécutant la cellule ci-dessous :

````python
content
````

Les symboles `\n` représentent le caractère de saut de ligne. Python va donc lire un fichier de manière brute en lisant aussi les sauts de ligne qui d'habitude sont « cachés » dans les éditeurs de texte.

## Écrire un fichier

Voici un exemple de code pour ouvrir un fichier et y écrire le contenu de la liste `animaux` :

````python
animaux = ["poisson", "abeille", "chat"]

with open("zoo2.txt", "w") as output_file:
    for animal in animaux:
        output_file.write(animal)
````

Remarques :

- Le mode `"w"` (pour *write*) est utilisé pour écrire dans un fichier.
- La méthode `.write()` permet d'écrire dans le fichier.

Ouvrez le fichier `zoo2.txt` en double-cliquant sur le nom du fichier dans l'explorateur de fichiers (à gauche).

Vous constatez que les noms des animaux sont tous écrits sur une seule ligne :

```none
poissonabeillechat
```

On voudrait plutôt écrire un nom d'animal par ligne.

Pour cela, il faut explicitement écrire les sauts de ligne avec le caractère `\n` dans le fichier :

````python
animaux = ["poisson", "abeille", "chat"]

with open("zoo3.txt", "w") as output_file:
    for animal in animaux:
        output_file.write(f"{animal}\n")
````

Ouvrez maintenant le fichier `zoo3.txt`. Vous devriez obtenir exactement ce que vous vouliez, à savoir, un animal par ligne.

## Format FASTA

En biologie, les séquences d'ADN ou de protéine sont souvent stockées dans des fichiers au format FASTA :

```none
>entête
séquence avec un nombre maximum de caractères par ligne
séquence avec un nombre maximum de caractères par ligne
séquence avec un nombre maximum de caractères par ligne
séquence avec un nombre maximum de caractères par ligne
séquence avec un nombre max
```

- La première ligne est toujours une ligne d'entête qui débute par le caractère `>` et qui contient des informations sur la séquence (identifiant, nom, organisme...).
- Les lignes suivantes sont constituées de la séquence répartie sur plusieurs lignes avec un nombre fixe de bases ou de résidus par ligne (sauf la toute dernière ligne).

Pour cette protéine, nous avons une ligne d'entête :

```none
>sp|P05106|ITB3_HUMAN Integrin beta-3 OS=Homo sapiens OX=9606 GN=ITGB3 PE=1 SV=2
```

et 13 lignes de séquence avec 60 résidus par ligne + 1 ligne avec les 8 derniers résidus.

On retrouve bien les 788 résidus de cette protéine :

$$
\rm 13 \; lignes \times 60 \; résidus + 1 \; ligne \times 8 \; résidus = 780 + 8 = 788
$$

Pour lire un fichier FASTA avec Python, on peut utiliser le code suivant qui va combiner la lecture d'un fichier, la répétition avec la boucle `for` et la prise de décision avec les tests `if` / `else` :

````python
with open("P05106.fasta", "r") as fasta_file:
    header = ""
    sequence = ""
    for line in fasta_file:
        if line.startswith(">"):
            header = line.strip()
        else:
            sequence += line.strip()
````

- La méthode `.startswith(">")` vérifie si une ligne débute par le caractère `>`
- La méthode `.strip()` supprime les retours en ligne en fin de ligne.
- L'instruction `sequence += line.strip()` est un raccouci pour `sequence = sequence + line.strip()`

On peut ensuite afficher l'entête : 

````python
print(header)
````

et les 10 premiers résidus de la séquence :

````python
print(sequence[0:10])
````

et sa taille :

````python
print(len(sequence))
````
